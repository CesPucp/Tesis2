```python
#Limpiar los abstracts
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re
lista_abs=[]

for ab in clean_df.Abstract:
    stop_words_english= set(stopwords.words('english'))
    stop_words_text=['background.','background','objective','in','this','paper','letter','article','work',
                    'so','far','aim','purpose', 'context','we', 'as','recent', 'according', 'objectives',
                    'objective.','and','ethnopharmacological', 'ethno-pharmacological','relevance','introduction']
    stop_words=stop_words_english.union(stop_words_text)
    
    filtered_tokens = []
    word_tokens = word_tokenize(ab)
    find=0
    for i in range(0,4):
        if (word_tokens[i]==':'):
            find=1
            break
    if(find):
        aux=word_tokens[i+1:]
    else:
        aux=word_tokens
        
    for token in aux:
        if re.search('[a-zA-Z0-9]', token):
            filtered_tokens.append(token.lower())
    filtered_token = [w for w in filtered_tokens if not w in stop_words]
    lista_abs.append(filtered_token)    
```

```python
lista_title=[]
for title in clean_df.Title:
    stop_words= set(stopwords.words('english'))
    filtered_tokens = []
    word_tokens = word_tokenize(title)
    for token in word_tokens:
        if re.search('[a-zA-Z0-9]', token):
            filtered_tokens.append(token.lower())
    filtered_token = [w for w in filtered_tokens if not w in stop_words]
    lista_title.append(filtered_token)  
```

```python
#Sacar la lista de los autores
#auxiliar
lista_autor=[]
lista_id_autor=[]
lista_id_publicacion=[]
lista_cantidad=[]
lista_coautores=[]
lista_publicacion=[]
lista_titulo=[]
lista_abstract=[]
lista_key=[]
lista_area=[]
lista_macro=[]

ind=0

#reordenar
for row in clean_df.Authors:
    #Extraer informacion
    sep=clean_df.AuthorsID[ind].split(';')
    name=clean_df.Authors[ind].split(',')
    key=clean_df.AuthorsKeywords[ind].split(';')
    
    cod_pub=clean_df.EID[ind]
    area=clean_df.Area[ind]
    macro=clean_df.MacroArea[ind]
    pub=lista_title[ind]+lista_abs[ind]
    size=len(sep)
    #print(size)
    for i in range(0,size-1):
        #si no esta el id autor se añade lista_id_autor
        if int(sep[i]) not in lista_id_autor:
            lista_id_autor.append(int(sep[i]))
            lista_autor.append(name[i])
            
            lista_id_publicacion.append([cod_pub])
            lista_publicacion.append(pub)
            lista_titulo.append(lista_title[ind])
            lista_abstract.append(lista_abs[ind])
            lista_cantidad.append(1)
            lista_coautores.append(sep[:-1])
            lista_area.append([area])
            lista_macro.append([macro])
            lista_key.append(key)
        else:
            #buscar el indice donde esta el autor
            indice=0
            count=0
            for cod in lista_id_autor:
                if cod==int(sep[i]):
                    indice=count
                    break
                count+=1
            #Añadir el codigo de la publicación en el autor
            lista_id_publicacion[indice].append(cod_pub)
            lista_publicacion[indice]+=pub
            lista_titulo[indice]+=lista_title[ind]
            lista_abstract[indice]+=lista_abs[ind]
            lista_cantidad[indice]+=1
            lista_coautores[indice]=lista_coautores[indice] + sep[:-1]
            lista_area[indice].append(area)
            lista_macro[indice].append(macro)
            lista_key[indice]+=key
    ind=ind+1
    if (ind%100)==0 :print(ind)
```

```python
def levenshteinDistance(s1, s2):
    if len(s1) > len(s2):
        s1, s2 = s2, s1

    distances = range(len(s1) + 1)
    for i2, c2 in enumerate(s2):
        distances_ = [i2+1]
        for i1, c1 in enumerate(s1):
            if c1 == c2:
                distances_.append(distances[i1])
            else:
                distances_.append(1 + min((distances[i1], distances[i1 + 1], distances_[-1])))
        distances = distances_
    return distances[-1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import re

def token (word):
    stop_words_spanish = set(stopwords.words('spanish'))
    stop_words_english = set(stopwords.words('english'))
    
    stop_words  = stop_words_spanish.union(stop_words_english)
    
    filtered_tokens = []
    #Tokenize and filter
    word_tokens = word_tokenize(word)
    for token in word_tokens:
        if re.search('[a-zA-Z0-9]', token):
            filtered_tokens.append(token.lower())
    
    filtered_token = [w for w in filtered_tokens if not w.lower() in stop_words]


    
    return filtered_token
```

```python
#kmeans Español
#Clustering
#tf idf
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt
#minimo=[0.001,0.005,0.01,0.015,0.018,0.02]
minimo=[0.001,0.005,0.01,0.015,0.018,0.02]
#clusters=[25,30,35,40,50]
clusters=[25,30,35,40,50]
lab=['25 Clusters','30 Clusters', '35 Clusters','40 Clusters','50 Clusters']

indice=0
for c in clusters:

   
    #print(c)
    promedio=[]
    for m in minimo:
        tfidf_vectorizer = TfidfVectorizer(max_df=0.2, max_features=20000000, min_df=m,
                                             use_idf=True, ngram_range=(1,3))
        tfidf_matrix_spanish = tfidf_vectorizer.fit_transform(data_spanish) 

        terms_spanish = tfidf_vectorizer.get_feature_names()
   
        print('Minimo:',m)
        print('Cantidad de terminos: ',len(terms_spanish))

        from sklearn.metrics.pairwise import cosine_similarity
        dist_spanish = 1 - cosine_similarity(tfidf_matrix_spanish)


        #kmeans
        from sklearn.cluster import KMeans

        num_clusters_spanish = c
        km = KMeans(n_clusters=num_clusters_spanish)
        km.fit(tfidf_matrix_spanish)
        results_spanish = km.predict(tfidf_matrix_spanish)

        #Metricas interna
        from sklearn.metrics import silhouette_samples, silhouette_score
        silhouette_avg_spanish = silhouette_score(tfidf_matrix_spanish, results_spanish)
        promedio.append(silhouette_avg_spanish)
        print('Numero de cluster:', num_clusters_spanish)
        print('Promedio de silhouette:',silhouette_avg_spanish)

        #Cluster
        for i in range(num_clusters_spanish):
            print("Cluster %d titles:" % i, end='')
            a=0
            cont=0
            for title in results_spanish:
                if(i==title):
                    #print (data_spanish[a], end=', ')
                    cont+=1
                    #print('')
                a+=1
            print(cont)

    plt.xlabel('Mínimo')
    plt.ylabel('Promedio Silhouette')
    print(indice)
    plt.plot(minimo,promedio,label=lab[indice])
    indice+=1
    
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.show()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
import matplotlib.pyplot as plt


tfidf_vectorizer = TfidfVectorizer(max_df=0.2, max_features=20000000, min_df=0.015,
                                     use_idf=True, ngram_range=(1,3))
tfidf_matrix_spanish = tfidf_vectorizer.fit_transform(data_spanish) 

terms_spanish = tfidf_vectorizer.get_feature_names()


print('Cantidad de terminos: ',len(terms_spanish))

from sklearn.metrics.pairwise import cosine_similarity
dist_spanish = 1 - cosine_similarity(tfidf_matrix_spanish)


#kmeans
from sklearn.cluster import KMeans

num_clusters_spanish = 35
km = KMeans(n_clusters=num_clusters_spanish)
km.fit(tfidf_matrix_spanish)
results_spanish = km.predict(tfidf_matrix_spanish)

#Metricas interna
from sklearn.metrics import silhouette_samples, silhouette_score
silhouette_avg_spanish = silhouette_score(tfidf_matrix_spanish, results_spanish)
promedio.append(silhouette_avg_spanish)
print('Numero de cluster:', num_clusters_spanish)
print('Promedio de silhouette:',silhouette_avg_spanish)

#Cluster
for i in range(num_clusters_spanish):
    print("Cluster %d titles:" % i, end='')
    a=0
    cont=0
    for title in results_spanish:
        if(i==title):
            #print (data_spanish[a], end=', ')
            cont+=1
            #print('')
        a+=1
    print(cont)
```

```python
from sklearn.cluster import KMeans
import numpy as np
from matplotlib import cm
from sklearn.metrics import silhouette_samples

X=tfidf_matrix_spanish
y_km=results_spanish
n_clusters=num_clusters_spanish

cluster_labels = np.unique(y_km)
n_clusters = cluster_labels.shape[0]
silhouette_vals = silhouette_samples(X, y_km, metric='euclidean')
y_ax_lower, y_ax_upper = 0, 0
yticks = []
for i, c in enumerate(cluster_labels):
    c_silhouette_vals = silhouette_vals[y_km == c]
    c_silhouette_vals.sort()
    y_ax_upper += len(c_silhouette_vals)
    color = cm.jet(float(i) / n_clusters)
    plt.barh(range(y_ax_lower, y_ax_upper), c_silhouette_vals, height=1.0, 
             edgecolor='none', color=color)

    yticks.append((y_ax_lower + y_ax_upper) / 2.)
    y_ax_lower += len(c_silhouette_vals)
    
silhouette_avg = np.mean(silhouette_vals)
print("For n_clusters =", n_clusters,
          "The average silhouette_score is :", silhouette_avg)
plt.axvline(silhouette_avg, color="red", linestyle="--") 

plt.yticks(yticks, cluster_labels + 1)
plt.ylabel('Cluster')
plt.xlabel('Silhouette coefficient')

plt.tight_layout()
# plt.savefig('./figures/silhouette.png', dpi=300)
plt.show()
```
